# MAXHEAP
- object :키 속성 지닌 요소 집합
- operation
1. h.is_empty() -> Boolean : 힙이 비어 있으면 True, 아니면 False
2. h.is_full() -> Boolean
3. h.push(element) : 요소 삽입
4. h.pop() -> element : 힙에서 최대 원소 삭제하며 반환

```python
class Element:
    def __init__(self,key):
        self.key = key
        
class MaxHeap:
    MAX_ELEMENTS = 100
    def __init__(self):
        self.arr = [None for i in range(self.MAX_ELEMENTS+1)]#None이 100개 있는 리스트
        self.heapsize = 0
        
    def is_empty(self):
        if self.heapsize ==0:
            return True
        return False
    
    def is_full(self):
        if self.heapsize >= self.MAX_ELEMENTS:
            return True
        #2 나눠줌
    def parent(self,idx):
        return idx >> 1
    
    #2를 곱해줌
    def left(self,idx):
        return idx << 1
    
    #2 곱하고 1 더해줌
    def right(self,idx):
        return (idx << 1)+1

    def push(self,item):
        if self.is_full():
            raise IndexError("the heap is full")
    
        #완전 이진 트리 유지하기 위해 마지막 원소의 다음 인덱스
        self.heapsize +=1
        cur_idx = self.heapsize
    
        #cur_idx가 루트가 아니며 & item의 key가 cur_idx 부모의 키보다 클 때까지는
        while cur_idx != 1 and item.key > self.arr[self.parent(cur_idx)].key:
            self.arr[cur_idx] = self.arr[self.parent(cur_idx)]
            #먼저 cur_idx 의 키 값을 부모의 키 값과 바꾸고
            cur_idx = self.parent(cur_idx)
            #cur_idx(인덱스) 번호 자체는 2로 나눠줌으로써 부모의 값을 할당함
            #item key = cur_idx 가 되거나 cur_idx = 1(루트) 가 되면 while문을 빠져나감
        self.arr[cur_idx] =item

        #Pop 연산 : 최대 힙은 pop 연산에서 힙에서 가장 큰 키 가진 요소를 삭제(가장큰키는 루트노드에)

    def pop(self):
        if self.is_empty():
            return None
    
        rem_elem = self.arr[1]#삭제후 반환될 원소
    
        #맨 마지막 위치한 원소 받아온후 힙사이즈 하나 줄이면
        # -> 완전 이진트리 특성 유지
        temp = self.arr[self.heapsize]
        self.heapsize -= 1
    
        cur_idx = 1# 루트에서 시작
        child = self.left(cur_idx)#루트의 왼쪽자식을 child default 설정
    
    
        while child <= self.heapsize:
            #오른쪽이 더크면 오른족으로 child갱신
            if child < self.heapsize and \
                    self.arr[self.left(cur_idx)].key < self.arr[self.right(cur_idx)].key:
                child = self.right(cur_idx)#오른쪽 자식이 왼쪽 자식 키보다 크니까 오른쪽을 child로 만들어줌
            if temp.key >= self.arr[child].key:
                break
#선택된 child와 temp 값을 바꿔주기    
            self.arr[cur_idx] = self.arr[child]
            cur_idx = child
            child = self.left(cur_idx)
            
        self.arr[cur_idx] = temp
        return rem_elem
```

```python
def print_heap(h):
    for i in range(1,h.heapsize+1):
        print("{}".format(h.arr[i].key),end=" ")
        
if __name__=="__main__":
    h = MaxHeap()
    h.push(Element(2))
    h.push(Element(14))
    h.push(Element(9))
    h.push(Element(11))
    h.push(Element(6))
    h.push(Element(8))
    
    print_heap(h)
    print(h.heapsize)
    while not h.is_empty():
        rem = h.pop()
        print(f"poped item is {rem.key}")
        print_heap(h)
```

# MinPriorityQueue
- object :키를 가진 원소 집합
- operation
1. is_empty() -> Boolean : 큐가 비어 있으면 True, 아니면 False
2. push(item) : 큐에 원소 item 삽입
3. pop() -> element : 큐에서 키 값 가장 작은 요소 삭제하며 반환
4. min() -> element
5. decrease_key(item,new_key) : 큐에 있는 요소 item의 키를 값이 줄어든 new_key 값으로 수정

```python
from heapq import heappush, heappop

class Element:
    def __init__(self, key, string):
        self.key=key
        self.data=string

class MinPriorityQueue:
    def __init__(self):
        self.heap=[]

    def is_empty(self):
        if not self.heap:
            return True
        return False

    def push(self, item):
        heappush(self.heap, (item.key, item.data))

    def pop(self):
        return heappop(self.heap)

    def min(self):
        return self.heap[0]

if __name__=="__main__":
    pq=MinPriorityQueue()

    pq.push(Element(2, "kim"))
    pq.push(Element(14, "park"))
    pq.push(Element(9, "choi"))
    pq.push(Element(11, "lee"))
    pq.push(Element(6, "yang"))
    pq.push(Element(8, "jang"))

    while not pq.is_empty():
        elem=pq.pop()
        print(f"key[{elem[0]}] : data[{elem[1]}]")
```
